```python
# Colab Cell 3 вЂ” Run the downloader (paste after the %%writefile cells above)
import sys
sys.path.insert(0, '.')

from src.ingestion.downloader import SECDownloader, FilingMetadata

COMPANIES = [
    {"ticker": "AAPL", "cik": "0000320193"},
    {"ticker": "MSFT", "cik": "0000789019"},
    {"ticker": "GOOGL", "cik": "0001652044"},
]

downloader = SECDownloader(output_dir="data/raw")
downloaded = []

for company in COMPANIES:
    print(f"\nв¬‡ Fetching {company['ticker']}...")
    metadata = downloader.get_latest_10k_url(
        cik=company["cik"],
        ticker=company["ticker"]
    )
    if metadata:
        path = downloader.download(metadata)
        if path:
            downloaded.append(metadata)
            print(f"  вњ… {metadata.ticker} | Filed: {metadata.filing_date}")

print(f"\nрџЋЇ Successfully downloaded {len(downloaded)}/3 filings")
```
